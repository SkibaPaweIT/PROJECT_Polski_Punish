```python
import csv
import pandas as pd
import statistics as st
import scipy.stats as ss

#Zad1
print("Zadanie 1")
file = pd.read_csv('zad1.csv')
kolumna = file['e_inc_rr_num_hi']
print(kolumna.sum())
print(kolumna.min())
print(kolumna.max())
print(kolumna.count())

#Zad2
print("Zadanie 2")
file2 = pd.read_csv('Wzrost.csv')
kolumna2 = file2.columns
wynik = list(map(float, kolumna2))
print("Srednia: " , st.mean(wynik))
print("Mediana: " , st.median(wynik))
print("Wariancja: " , st.variance(wynik))
print("Odchylenie std: " , st.pstdev(wynik))
#Odchylenie std to Wariacja podniesiona do kwadratu

#Zad3 
# cars.csv from https://perso.telecom-paristech.fr/eagan/class/igr204/datasets

print("Zadanie 3")
file3 = pd.read_csv('sample_data/california_housing_train.csv')
kolumna3 = file3['longitude']
print(kolumna3.kurtosis())
print(kolumna3.skew())
print(kolumna3.describe())


#zad 4
print("Zadanie 4")
file4 = pd.read_csv('brain_size.csv' , sep=';')

#srednia z VIQ
kolumna4 = file4['VIQ']
print(kolumna4.mean())

#Ile kobiet i mezczyzn
data_pogrupowane = file4.groupby('Gender')
for gender, value in data_pogrupowane['Gender']:
  print((gender,value.count()))


# Dla wszystkich
pd.plotting.scatter_matrix(file4[['VIQ' , 'PIQ' , 'FSIQ']])

# Tylko dla kobiet
for gender, value in data_pogrupowane['Gender']:
  if(gender == 'Female'):
    print('Kobiety:')
    pd.plotting.scatter_matrix(file4[['VIQ' , 'PIQ' , 'FSIQ']])







```
